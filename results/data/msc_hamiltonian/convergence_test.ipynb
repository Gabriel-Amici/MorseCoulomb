````python
import numpy as np
import matplotlib.pyplot as plt
import json
````

````python
files = ["a--2-0--L--5000--N--5001.json",
         "a--2-0--L--5000--N--10001.json", 
         "a--2-0--L--5000--N--20001.json"]

for file in files:
    fh = open(file, "r")
    data = json.load(fh)
    fh.close()

    print(data["results"]["bound_states"])

    position_grid = np.linspace(data["position_grid"]["postion_grid_start"], 
                                data["position_grid"]["postion_grid_start"]+data["position_grid"]["position_grid_extension"],
                                data["position_grid"]["position_grid_size"])
    plt.plot(position_grid, np.array(data["results"]["msc_states"])[:, 200], 
             label=fr"$L={data["position_grid"]["position_grid_extension"]}, N={data["position_grid"]["position_grid_size"]}, n=20$")
plt.xlim(4800, 5000)
plt.legend()
plt.show()
````

````python
files = ["a--2-0--L--5000--N--5001.json",
         "a--2-0--L--5000--N--10001.json", 
         "a--2-0--L--5000--N--20001.json"]

N = []
E0 = []
E1 = []
E2 = []

for file in files:
    fh = open(file, "r")
    data = json.load(fh)
    fh.close()

    print(data["results"]["bound_states"])

    N.append(data["position_grid"]["position_grid_size"])
    E0.append(data["results"]["msc_energies"][0])
    E1.append(data["results"]["msc_energies"][1])
    E2.append(data["results"]["msc_energies"][2])

plt.plot(N, E0, label="$E_0$", marker=".")
plt.plot(N, E1, label="$E_1$", marker=".")
plt.plot(N, E2, label="$E_2$", marker=".")

plt.legend()
plt.grid()
plt.show()
````

Conclusão: para $\alpha = 2$ e $L = 5000$, $N = 5001$ basta. $E(N) = cte$, número de pontos no grid não afeta consideravelmente a energia
