```python
colors = ["red", "green", "blue", "orange", "pink"]

alpha = 0.5
L = 1000
N = 5001
n = 100
total_time = 1000
delta_t = 0.01
F_0 = 0.01

delta_t_array = [50, 10, 5, 1]

for j in range(4):
    delta_t = delta_t_array[j]

    data = converge_test(alpha, L, N, n, total_time, delta_t, F_0)
    coefficients_history = np.array(data["Simulation Results"]["coeficcients_history"])
    time_grid = np.linspace( data["time_grid"]["lower_bound"], data["time_grid"]["lower_bound"] + data["time_grid"]["extension"], data["time_grid"]["points"] )

    ax[0][j].set_title(r"$\Delta t={}$".format(delta_t))
    for k in range(5):
        ax[0][j].plot(time_grid, coefficients_history[:, k], label="$n={}$".format(k+1), c=colors[k])


delta_t_array = [0.5, 0.1, 0.01, 0.001]

for j in range(4):
    delta_t = delta_t_array[j]
    spacing = int(1/delta_t)

    data = converge_test(alpha, L, N, n, total_time, delta_t, F_0)
    coefficients_history = np.array(data["Simulation Results"]["coeficcients_history"])
    time_grid = np.linspace( data["time_grid"]["lower_bound"], data["time_grid"]["lower_bound"] + data["time_grid"]["extension"], data["time_grid"]["points"] )

    ax[1][j].set_title(r"$\Delta t={}$".format(delta_t))
    for k in range(5):
        ax[1][j].plot(time_grid[::spacing], coefficients_history[::spacing, k], label="$n={}$".format(k+1), c=colors[k])
    
```
