### Este projeto estuda as propriedades de um potencial inédito chamado Morse-soft-Coulomb

A tarefa de estudar a dinâmica de potenciais unidimensionais é de grande importância...

$$
V_{MsC}(r) = \begin{cases}-\dfrac{1}{ \sqrt{r^2 + \alpha^2}} , ~~r > 0 \\~\\D(e^{-2\beta r} -2e^{-\beta r}) , ~~r \leqslant 0.\end{cases}
$$

```python
from emerald.potentials.msc_potential import MsC_potential
from emerald.potentials.coulomb_potential import C_potential
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
r_array1 = np.linspace(-2, 10, 200)
r_array2 = np.linspace(0.001, 10, 200)
for alpha in [2, 1, 0.5, 0.1]:
    plt.plot( r_array1, [ MsC_potential(alpha, r) for r in r_array1 ], lw=1, label="$\\alpha = {}$".format(alpha) )
plt.plot(r_array2, [C_potential(r) for r in r_array2], label="$\\alpha = 0$ (Coulomb)", ls="dashed", c="black")
plt.xlim(-2, 10) ; plt.ylim(-3, 2)
plt.legend() ; plt.show()
```
